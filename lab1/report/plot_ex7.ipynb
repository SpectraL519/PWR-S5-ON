```python
import sys

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ex7.csv')
df.head()
```

```python
# plot derivative values and approximations
fig = plt.figure(figsize=(6, 4))
plt.title("derivative approximation")

plt.scatter(
    df["n"], df["df"],
    label="derivative", c="blue", s=5
)
plt.scatter(
    df["n"], df["df_approx"],
    label="approximation", c="red", s=5
)

plt.xlabel("n")
plt.legend()
plt.show();
```

```python
# plot absolute error
fig = plt.figure(figsize=(6, 4))
plt.title("derivative approximation - absolute error")

plt.scatter(
    df["n"], df["approx_error"],
    c="red", s=5
)

plt.xlabel("n")
plt.ylabel("absolute error")
plt.legend()
plt.show();
```
